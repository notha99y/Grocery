# Getting the Data

```python
def requestForProducts(val, page):
    r = get(
        'https://api.redmart.com/{}/catalog/search?category={}&pageSize={}&page={}'.format(
            VERSION, val, PAGESIZE, page))
    data = json.loads(r.text)
    return data
```
